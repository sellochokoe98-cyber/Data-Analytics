```python
import pandas as pd
```

```python
df_chocolate = pd.read_csv(r"C:\Users\sello\Downloads\Chocolate Sales.csv")

df_chocolate.head()
```

```python
### ("Total sales amount")
total_sales = df_chocolate['Amount'].sum()
total_sales
```

```python
### ("Total boxes shipped")
total_boxes = df_chocolate['Boxes Shipped'].sum()
total_boxes
```

```python
### ("Sales by country")
sales_by_country = df_chocolate.groupby('Country')['Amount'].sum()
sales_by_country
```

```python
### ("Top sales person")
sales_by_person = df_chocolate.groupby('Sales Person')['Amount'].sum().sort_values(ascending=False)
sales_by_person.head()
```

```python
### ("Most sold product")
product_sales = df_chocolate.groupby('Product')['Amount'].sum().sort_values(ascending=False)
product_sales
```

```python
### ("Average sales amount")
average_sales = df_chocolate['Amount'].mean()
average_sales
```

```python
### ("Monthly sales trend")
monthly_sales = df_chocolate.groupby(df_chocolate['Date'].dt.month)['Amount'].sum()
monthly_sales
```

```python
### ("Boxes shipped per product")
boxes_by_product = df_chocolate.groupby('Product')['Boxes Shipped'].sum()
boxes_by_product
```

```python
###("Simple bar chart (sales by country)")

import matplotlib.pyplot as plt

df_chocolate.groupby('Country')['Amount'].sum().plot(kind='bar')
plt.title('Sales by Country')
plt.xlabel('Country')
plt.ylabel('Sales Amount')
plt.show()
```
