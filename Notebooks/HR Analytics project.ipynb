```python
# Count total duplicate rows
num_duplicates = df_HR.duplicated().sum()
print(f"Total duplicate rows: {num_duplicates}")
```

```python
duplicate_rows = df_HR[df_HR.duplicated()]
print("Duplicate rows:")
print(duplicate_rows)
```

```python
df_HR_clean = df_HR.drop_duplicates()
print("Shape after removing duplicates:", df_HR_clean.shape)
```

```python
missing_values = df_HR.isnull().sum()
print("Missing values per column:")
print(missing_values)
```

```python
unique_values = df_HR.nunique()
print("Unique values per column:")
print(unique_values)
```

```python
age_outliers = df_HR[(df_HR['Age'] < 18) | (df_HR['Age'] > 65)]
print("Rows with age < 18 or > 65:")
print(age_outliers)
```
